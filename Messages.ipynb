```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash")

model.invoke("explain langchaning in few words")
```

Test prompts - A simple message to the model 
Message prompts - List of messages to the llm by providing list of message objects

System msg - Instruction to model how it should be have  , Human msg - User input , AI msg  - reponse generated by the model , Tool msg - reponse returned from the tools

```python
from langchain.messages import SystemMessage , HumanMessage 

message = [ SystemMessage("You are an excellent poet") , 
           HumanMessage(" write a poem on AI")]

response =model.invoke(message)
response.content
```
