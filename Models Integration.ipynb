### Model Integration with gemini


```python
import os 
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model="google_genai:gemini-2.5-flash")

response = model.invoke("Hello !! How are you ?")

response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
response = model.invoke("Hello !! How are you?")
response.content
```

### Stream()

For longer text generation we have to wait longer period until llm generates the whole answer and to answer , hence we can use stream to chunk the answers and provide it to the end user

```python
for chunk in model.stream("Write a 100 word essay about the importance of Prayer"):
    print(chunk.text , end = "|" , flush = True)
```

### Batch 

Paralell computing for multiple texts

```python
responses = model.batch(["What is your name ? " , "What is your age" , "Where do you stay"] , config = {"max_concurrency" : 5})
for response in responses:
    print(response)
```
