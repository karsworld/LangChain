###Tools 

```python
from langchain.tools import tool
from langchain.chat_models import init_chat_model

@tool
def get_whether(location : str):
    """This function is used to get the whether"""
    f"The whther in {location} is sunny"

model = init_chat_model("google_genai:gemini-2.5-flash")
 
model_with_tool = model.bind_tools([get_whether])

response = model_with_tool.invoke("What is the whether in boston")

for tool_call in response.tool_calls:
    print(f"Tool_name is {tool_call['name']}")
    print(f"Args are {tool_call['args']}")

```
